```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('adult.data.csv')
```

```python
df.head(5)
```

```python
df.describe()
```

```python
#1. How many men and women are represented in dataset?

df['sex'].value_counts()
```

```python
#2. What is the average age (age feature) of women?

df[df['sex'] == 'Female']['age'].mean()
```

```python
#3. What is the percentage of German citizens (native-country feature)?

df['native-country'].value_counts(normalize=True).get('Germany')*100
```

```python
'''4. What are the mean and standard deviation of age for those who earn more
than 50K per year (salary feature) and those who earn less than 50K per
year?'''

df.groupby(['salary'])['age'].describe(percentiles=[])
```

```python
#5. Is it true that people who earn more than 50K have at least high school
#education? (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc,
#Masters or Doctorate feature)

pd.crosstab(df['salary'], df['education-num'], margins=True)
```
